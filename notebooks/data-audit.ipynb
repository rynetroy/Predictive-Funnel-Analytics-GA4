```python
import os
import pandas as pd
from ydata_profiling import ProfileReport

os.makedirs("reports", exist_ok=True)

customers = pd.read_csv('../data/customers.csv')
campaigns = pd.read_csv('../data/campaigns.csv')
events = pd.read_csv('../data/events.csv')
products = pd.read_csv('../data/products.csv')
transactions = pd.read_csv('../data/transactions.csv')

ProfileReport(customers, title="Customers Report", explorative=True).to_file("reports/customers_report.html")
ProfileReport(campaigns, title="Campaigns Report", explorative=True).to_file("reports/campaigns_report.html")
ProfileReport(products, title="Products Report", explorative=True).to_file("reports/products_report.html")

events_sample = events.sample(min(10000, len(events)), random_state=42)
transactions_sample = transactions.sample(min(10000, len(transactions)), random_state=42)

ProfileReport(events_sample, title="Events Sample Report", explorative=True).to_file("reports/events_sample_report.html")
ProfileReport(transactions_sample, title="Transactions Sample Report", explorative=True).to_file("reports/transactions_sample_report.html")

print("Reports created in the reports/ folder.")
```
